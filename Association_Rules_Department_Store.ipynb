```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('June.xlsx')
```

```python
#行列計算總數
#basket = pd.pivot_table(df,index=['date_card_number'],columns=['counter'],values=['frequency'],aggfunc=[np.sum],fill_value=0,margins=1)
basket = pd.pivot_table(df,index=['date_card_number'],columns=['counter'],values=['frequency'],aggfunc=[np.sum]).fillna(0)
basket
```

```python
#將計數轉換為0和1
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
#將計數轉換為0和1
basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.0005, use_colnames=False)
#use_colnames=True表示使用名字，預設的False使用列數
frequent_itemsets
```

```python
frequent_itemsets.sort_values(by='support',ascending=False,inplace=True)
#依support排序(排序)
frequent_itemsets
```

```python
frequent_itemsets2 = frequent_itemsets[frequent_itemsets.itemsets.apply(lambda x: len(x)) >= 2]
# 篩選頻繁項集>= 2
frequent_itemsets2
```

```python
association_rule = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
# metric可以有很多的度量選項，返回的表列名都可以作為引數
association_rule
```

```python
association_rule.sort_values(by='confidence',ascending=False,inplace=True)
#依by=support,confidence...排序
```

```python
frequent_itemsets.to_csv('frequent_itemsets.csv',encoding='utf_8_sig')
association_rule.to_csv('association_rule.csv',encoding='utf_8_sig')
```
